```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cars = os.path.join("..","electric_cars","vehicle-fuel-type-count-by-zip-code.csv")
geocodes_raw = os.path.join("..","electric_cars","fips_zip_x.csv")
cars_data = pd.read_csv(cars)
geocodes = pd.read_csv(geocodes_raw)
cars_data.head()
```

```python
#getting the number of zips per county, by grouping by zip and then counting number of counties
n_zip_county = geocodes.groupby('ZCTA5').count().sort_values(['COUNTY'],ascending = False).reset_index()

#creating weights by doing 1/n
n_zip_county['Weights'] = 1/n_zip_county['COUNTY']
n_zip_county
```

```python
#getting types of Fuel from data
cars_description = cars_data["Fuel"].unique()
print(cars_description)
```

```python
#renaming zip code for merge
n_zip_county =n_zip_county.rename(columns={"ZCTA5":"Zip Code"})
cars_data =cars_data.rename(columns={'Zip Code': 'Zip Code'})
n_zip_county
```

```python
#merging weights/number of car regristrations by zip code into fips code
cars_merge_county = pd.merge(cars_data, n_zip_county, how="inner", on=["Zip Code"])
cars_merge_county.head().sort_values(['Weights'])
```

```python
#checking weights
cars = cars_merge_county.dropna().sort_values(['Weights'])
cars.head()
```

```python
#reducing extra rows
cars_cleaned = cars.drop(columns=['Date', 'Model Year', 'Make', 'Duty', 'Unnamed: 0', 'STATE'])

#multiplying the weights by number of vehicles
cars_cleaned['Adj Veh'] = cars_cleaned['Vehicles']*cars_cleaned['Weights']

cars_cleaned.head().sort_values(['Weights'])
```

```python
#creating unique DF for each type of fuel type
battery_electric = cars_cleaned.loc[cars_cleaned['Fuel']=="Battery Electric"]
plug_in = cars_cleaned.loc[cars_cleaned['Fuel']=="Plug-in Hybrid"]
hybrid = cars_cleaned.loc[cars_cleaned['Fuel']=='Hybrid Gasoline']

#finding the sum of each fuel type
batt_car_total = battery_electric['Vehicles'].sum()
hybrid_total=hybrid['Vehicles'].sum()
plug_in_total = plug_in['Vehicles'].sum()



#creating bar chart for all vehicle types less gasoline, based on adjusted vehicles
fuel = cars_cleaned[['Fuel','Adj Veh']].groupby('Fuel').sum().reset_index()
fuel_no_gas = fuel.loc[(fuel['Fuel'] != 'Gasoline')]
x_axis = np.arange(len(fuel_no_gas))
tick_locations = [value for value in x_axis]

plt.bar(x_axis, fuel_no_gas['Adj Veh'])
plt.xticks(tick_locations, fuel_no_gas['Fuel'], rotation = 'vertical')
plt.show()

```

```python
#mapping adjusted vehicles back into cars_merge_county to get the county ID back
cars_county_veh = pd.merge(cars_merge_county, cars_cleaned,
                          how = 'left' on 'Zip Code')
cars_county_veh
```
